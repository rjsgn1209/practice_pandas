## Boolean indexing
- boolean array를 index로 하여 True에 해당하는 값만을 추출..
- array의 길이와 boolean array의 길이가 같아야 한다..

```python
ta = np.array([1,4,5,6,3,5,6])
```

```python
# ta에서 5 이상인 값만 출력하시오...
```

```python
for i in ta:
    if i >= 5:
        print(i)
```

```python
ta >= 5  
```

```python
ta[ta >= 5] # 데이터 분석시 속도가느려 for문을 잘 사용하지않
```

```python
# ta에서 짝수만 출력하시오..
```

```python
ta[ta % 2 == 0]
```

```python
# 1 ~ 100 사이의 정수에서 20 보다 크고 30 보다 작은 3의 배수를 출력하시오.. 
# => boolean indexing을 사용..
```

```python
tv = np.arange(1, 101)
tv = tv[30 > tv]
tv = tv[tv > 20]
tv = tv[tv%3 == 0]
tv[(30>tv) & (tv>20) & (tv%3 == 0)]
```

```python
tm = np.array([[1,2,3,4], [5,6,8,2]])
tm2 = np.array([[1],[8]])
tm2
```
